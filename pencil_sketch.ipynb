**Importing Libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from google.colab.patches import *
```

**Loading and Showing image**

```python
plt.figure(figsize=(15,8))
img= cv2.imread("/content/cat.jpg")
cv2_imshow(img)
```

Converting the image into RGB format

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2_imshow(img_rgb)
```

Converting the image into Greyscale

```python
g_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(g_image)
```

 Converting the image into Inverted Image

```python
inverted_img=255-g_image
cv2_imshow(inverted_img)
```

Converting the image into Gaussian Blur


```python
b=cv2.GaussianBlur(inverted_img,(21,21),0)
cv2_imshow(b)
```

Converting the image into Inverted Blurred

```python
in_blurred=255-b
cv2_imshow(in_blurred)
```

 Converting the image into Sketch

```python

pencil_sketch=cv2.divide(g_image,in_blurred,scale=256.0)
cv2_imshow(pencil_sketch)
```
