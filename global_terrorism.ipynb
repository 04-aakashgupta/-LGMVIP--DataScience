## Data Cleaning

### Removing unnecessary Columns

```python
gt=gt[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
```

```python
gt.columns
```

### Rename Column Name

```python
gt.rename(columns={"iyear":"year","imonth":"month","iday":"day","country_txt":"country",
                "region_txt":"region","provstate":"State","city":"city",
                "attacktype1_txt":"attacktype","targtype1_txt":"targettype",
                 "gname":"groupname","weaptype1_txt":"weapontype",
                "nkill":"killed","nwound":"wounded"},inplace=True)
```

```python
gt.columns
```

### Removing Duplicate Rows

```python
gt.duplicated().sum()
```

```python
gt.shape
```

```python
gt=gt.drop_duplicates()
```

```python
gt.duplicated().sum()
```

```python
gt.shape
```

### Filling Null Values 

```python
gt.isnull().sum()
```

```python
gt['latitude']=gt['latitude'].fillna(0)
gt['longitude']=gt['longitude'].fillna(0)
gt['killed']=gt['killed'].fillna(0)
gt['wounded']=gt['wounded'].fillna(0)
```

```python
gt.isnull().sum()
```

```python
gt.drop(['location','summary','addnotes','motive','weapontype'],inplace
        =True,axis=1)
```

```python
print(gt.columns)
```
